# NWS Standard Color Curves (2018)

```python
from paint.standard2 import *
```

```python
cm = [cm_wind, cm_tmp, cm_rh, cm_cloud, cm_dpt, cm_pcp, cm_pop, cm_snow, cm_wave_height]
for c in cm:
    print(c)
    z = c()
    z.display()
    if c == cm_pop:
        c(ptype='ice').display()
        c(ptype='snow').display()
```

# Demo a Colorbar

```python
z = cm_tmp()

ls = np.linspace(0, 3*np.pi, 100)
xx, yy = np.meshgrid(ls, ls)
ax = plt.axes()
img = ax.pcolormesh(xx, yy, np.sin(xx) * np.cos(yy)*50,
                    **z.cmap_kwargs)
plt.colorbar(img, **z.cbar_kwargs)
```

```python
z = cm_wind()
z.cbar_kwargs['spacing'] = 'uniform'

ls = np.linspace(0, 3*np.pi, 100)
xx, yy = np.meshgrid(ls, ls)
ax = plt.axes()
img = ax.pcolormesh(xx, yy, np.sin(xx) * np.cos(yy)*75,
                    **z.cmap_kwargs)
plt.colorbar(img, **z.cbar_kwargs)
```
